```python
import pandas as pd
import plotly.express as px


# =========================================================
# Load Dataset
# =========================================================

df = pd.read_csv("/Users/sahilsantoshdangat/Downloads/gapminder.csv")

print(f"Loaded: {len(df)} rows")
print(df.head())


# =========================================================
# Explore Dataset
# =========================================================

print("\nINFO")
print(df.info())

print("\nYears:")
print(sorted(df["year"].unique()))

print("\nContinents:")
print(df["continent"].unique())

print("\nSummary Statistics")
print(df.describe().round(1))


# =========================================================
# Task 1 — Scatter: Life Expectancy Change Over Time
# =========================================================
# Show GDP per capita vs Life Expectancy
# for 2002 and 2007
# using ONE continent only
# =========================================================

# Choose one continent (except Africa)
continent_choice = "Europe"

# Filter data
task1_df = df[
    (df["continent"] == continent_choice) &
    (df["year"].isin([2002, 2007]))
]

print("\nTask 1 Dataset Preview")
print(task1_df.head())


# Create scatter chart
fig1 = px.scatter(
    task1_df,
    x="gdpPercap",
    y="lifeExp",
    color="year",
    hover_name="country",

    # Rule: Log scale
    log_x=True,

    # Minimal colours
    color_discrete_sequence=[
        "#b0b0b0",   # grey
        "#1f77b4"    # blue
    ],

    title="European Countries Increased Life Expectancy as GDP Grew (2002 → 2007)",

    labels={
        "gdpPercap": "GDP per Capita (Log Scale)",
        "lifeExp": "Life Expectancy",
        "year": "Year"
    }
)

# Improve style
fig1.update_traces(
    marker=dict(
        size=10,
        opacity=0.8
    )
)

fig1.update_layout(
    template="simple_white",
    title_font_size=20
)

# Add annotation
fig1.add_annotation(
    x=25000,
    y=82,
    text="Most countries moved upward and right,<br>showing economic growth and better health outcomes.",
    showarrow=True,
    arrowhead=2
)

fig1.show()


# =========================================================
# Task 2 — Bubble Chart: Tell a Story
# =========================================================
# x = GDP per capita (log scale)
# y = life expectancy
# size = population
# highlight ONE continent
# others grey
# =========================================================

# Filter year 2007
df_2007 = df[df["year"] == 2007].copy()

# Choose continent to highlight
highlight_continent = "Asia"

# Create highlight column
df_2007["highlight"] = df_2007["continent"].apply(
    lambda x: highlight_continent if x == highlight_continent else "Other Continents"
)

print("\nTask 2 Dataset Preview")
print(df_2007.head())


# Create bubble chart
fig2 = px.scatter(
    df_2007,

    x="gdpPercap",
    y="lifeExp",

    size="pop",

    color="highlight",

    hover_name="country",

    # Rule: log scale
    log_x=True,

    # Rule: set size max
    size_max=60,

    title="Asian Countries Show Large Economic Differences but Strong Life Expectancy Outcomes (2007)",

    labels={
        "gdpPercap": "GDP per Capita (Log Scale)",
        "lifeExp": "Life Expectancy",
        "pop": "Population"
    },

    # Grey + highlight method
    color_discrete_map={
        highlight_continent: "#d62728",
        "Other Continents": "#c7c7c7"
    }
)

# Improve appearance
fig2.update_traces(
    marker=dict(
        opacity=0.75,
        line=dict(width=0.5, color="white")
    )
)

fig2.update_layout(
    template="simple_white",
    title_font_size=20,
    showlegend=False
)

# Add annotation
fig2.add_annotation(
    x=5000,
    y=72,
    text="Asia includes both developing and highly developed economies,<br>creating a wide spread in GDP per capita.",
    showarrow=True,
    arrowhead=2
)

fig2.show()
```
